```python
# Import dependencies
import pandas as pd
```

```python
# Import csv
vine_df = pd.read_csv('vine_table.csv')
```

```python
# print Vine DataFrame
vine_df
```

```python
# Filter reviews on total votes greater than 20
vine_votes_df = vine_df[vine_df['total_votes'] >=20]
```

```python
# Print DataFrame
vine_votes_df
```

```python
# Calculate percentage of helpful votes based on total votes
votes_calc = [vine_votes_df["helpful_votes"]/vine_votes_df["total_votes"] * 100]
votes_calc
```

```python
# Create additional column and add calculation
vine_votes_df.insert(6, "votes_calc", (vine_votes_df["helpful_votes"]/vine_votes_df["total_votes"] * 100), True)
```

```python
vine_votes_df
```

```python
# Create new DataFrame with filtered results based on percentage over 50%
filtered_vine_votes_df = vine_votes_df[vine_votes_df['votes_calc'] >=50]
```

```python
filtered_vine_votes_df
```

```python
# Create new DataFrame with filtered results over 50% and review is paid
vine_filtered_vine_votes_df = filtered_vine_votes_df[filtered_vine_votes_df['vine'] == 'Y']
vine_filtered_vine_votes_df
```

```python
# Create new DataFrame with filtered results over 50% and review is not paid
no_vine_filtered_vine_votes_df = filtered_vine_votes_df[filtered_vine_votes_df['vine'] == 'N']
no_vine_filtered_vine_votes_df
```
